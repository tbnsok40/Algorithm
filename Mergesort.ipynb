```python
# 합병정렬 다시 푼 것 191204

def merge(list1, list2):
    
    new = []
    
    i = 0
    j = 0
    
    while (i != len(list1)) & (j != len(list2)):
        
        if list1[i] > list2[j]:
            new.append(list2[j])
            j += 1
            
        elif list1[i] < list2[j]:
            new.append(list1[i])
            i += 1
    
    if i == len(list1):
        new = new + list2[j:]
    else:
        new = new + list1[i:]
        
    return new
    

    
def merge_sort(lists):
    if len(lists) < 2:
        return lists
    
    else:
        m = len(lists) // 2
        left_list = lists[:m]
        right_list = lists[m:]

    return merge(merge_sort(left_list), merge_sort(right_list))


# 같은 원소 있을 때는 에러발생
# 원인 찾아보기


print(merge_sort([1, 3, 5, 7, 9, 11, 13, 11]))
print(merge_sort([28, 13, 9, 30, 1, 48, 5, 7, 15]))
print(merge_sort([2, 5, 6, 7, 1, 2, 4, 7, 10, 11, 4, 15, 13, 1, 6, 4]))
```

```python
def merge(list1, list2): # 두개로 쪼개진(Divided) 리스트를 합친다(Conquere).
    
    i = 0
    j = 0

    # 정렬된 항목들을 담을 리스트
    merged_list = []

    # list1과 list2를 돌면서 merged_list에 항목 정렬
    while i < len(list1) and j < len(list2):
        if list1[i] > list2[j]:
            merged_list.append(list2[j])
            j += 1
        else:
            merged_list.append(list1[i])
            i += 1

    # list2에 남은 항목이 있으면 정렬 리스트에 추가
    if i == len(list1):
        merged_list += list2[j:]

    # list1에 남은 항목이 있으면 정렬 리스트에 추가
    elif j == len(list2):
        merged_list += list1[i:]

    return merged_list 
    #쪼개진 리스트에서 축출한 원소들을 차례로 merged_list에 넣었으니 이걸을 return한다.


# 합병 정렬
def merge_sort(my_list):

    if len(my_list) < 2:
        return my_list
    
    
    # 하나의 리스트를 인자로 받는다.
    # 이걸 나눠야 한다 반으로. 
    # 언제까지?
    # 반으로 나눈 리스트의 순서를 상관하지 않도록, 아예 길이가 1인 원소로 나눠버린다.
    # 그걸 merge함수의 인자로 넣어버린다.
    # merge함수 작동
    # Boom bamm
    
    mid_ind = len(my_list) // 2
    left_half = my_list[:mid_ind]
    right_half = my_list[mid_ind:]
    
    return merge(merge_sort(left_half), merge_sort(right_half))
    
    # 리스트를 반으로 나누는 함수를 재귀적으로 호출하여 원소의 개수가 하나 이하일 때까지 나눈다(divide).

    
        
    
    # when length of each list is 1, merge fucntion initiate.    

# 테스트
print(merge_sort([1, 3, 5, 7, 9, 11, 13, 11]))
print(merge_sort([28, 13, 9, 30, 1, 48, 5, 7, 15]))
print(merge_sort([2, 5, 6, 7, 1, 2, 4, 7, 10, 11, 4, 15, 13, 1, 6, 4]))
```

### merge_sort(): 절반으로 나눠준다. 언제까지? 리스트 내의 원소가 하나 이하일 때 까지.
### merge(list1, list2): 위에서 나눠준 리스트를 새로운 리스트에 순서대로(오름차순) append 해준다. 언제까지? 한쪽 리스트의 원소가 동이 날때까지. 한쪽이 동이 나면 다른쪽의 리스트는 다 갖다박아
